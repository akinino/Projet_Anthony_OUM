*Part 2: Calculating centrality metrics using Python*

Let’s consider a network with 10 vertices and edges distributed as following: [(7,2), (2,3), (7,4), (4,5), (7,3),
(7,5), (1,6),(1,7),(2,8),(2,9)]
1. Create the associated graph by using networkx package.

```python
edges = [(7,2), (2,3), (7,4), (4,5), (7,3), (7,5), (1,6),(1,7),(2,8),(2,9)]
G = nx.Graph()
G.add_edges_from(edges)

def print_centrality(centrality_type, centrality_dict):
    print(f"{centrality_type} Centrality:")
    for node, centrality in centrality_dict.items():
        print(f"Node {node}: {centrality}")
```

2. Calculate its degree centrality

```python
#La centralité degré quantifie le nombre de liens qu'un nœud possède.
degree_centrality = nx.degree_centrality(G)
print_centrality("La centralité degré quantifie le nombre de liens qu'un nœud possède. \nUne valeur proche de 1 indique un nœud fortement connecté et influent, tandis qu'une valeur proche de 0 suggère un nœud moins influent et moins connecté dans le réseau. \nDegree", degree_centrality)
```

3. Calculate its betweenness centrality

```python
#La centralité intermédiarité évalue le degré auquel un nœud se situe sur les chemins reliant d'autres nœuds.
betweenness_centrality = nx.betweenness_centrality(G)
print_centrality("La centralité d'intermédiarité mesure à quel point un nœud se trouve sur les chemins qui relient d'autres nœuds. \nUne valeur plus élevée indique que ce nœud joue un rôle central dans la communication entre les autres nœuds, tandis qu'une valeur plus basse suggère qu'il est moins central dans ce processus.\nBetweenness", betweenness_centrality)
```

4. Calculate its closeness centrality


```python
#La centralité de proximité évalue la distance à laquelle un nœud se trouve de tous les autres nœuds dans le graphe.
closeness_centrality = nx.closeness_centrality(G)
print_centrality("La centralité de proximité mesure la distance entre un nœud donné et tous les autres nœuds dans le graphe. \nUne valeur plus élevée indique que le nœud est plus proche de la majorité des autres nœuds, tandis qu'une valeur plus basse indique qu'il est plus éloigné des autres nœuds.\nCloseness", closeness_centrality)
```

*Part 3: Fraud analytics*

Let's look at how we can use SNA to detect fraud.


The word homophily has been coined to represent the effect human social network has on a person. Extending
this concept, a homophilic network is a group of people who are likely to be associated with each other due to
some common factor; for example, having the same origin or hobbies, being part of the same gang or the same
university, or some combination of other factors.


If we want to analyze fraud in a homophilic network, we can take advantage of the relationships between the
person under investigation and other people in the network, whose risk of involvement in fraud has already been
carefully calculated. Flagging a person due to their company is sometimes also called guilt by association.


In an effort to understand the process, let's first look at a simple case.
1. Create a network with nine vertices and eight edges distributed as following: [(7,2), (2,3), (7,4), (4,5),
(7,3), (7,5),(1,6),(1,7),(2,8),(2,9)]

2. In this network, four of the vertices are known fraud cases and are classified as **fraud (F)**. Five of the
remaining people have no fraud-related history and are classified as **non-fraud (NF)**. Define nodes
[1,4,3,8,9] as **NF** and [2,5,6,7] as **F**.

```python
# Définit les nœuds frauduleux et non-frauduleux
fraud_nodes = [2, 5, 6, 7]
non_fraud_nodes = [1, 4, 3, 8, 9]
```

##*The watchtower fraud analytics methodology*

The previous, simple fraud analytics technique has the following two limitations:
- It does not evaluate the importance of each vertex in the social network. A connection to a hub that is
involved in fraud may have different implications than a relationship with a remote, isolated person.
- When labeling someone as a known case of fraud in an existing network, we do not consider the severity
of the crime.
The watchtower fraud analytics methodology addresses these two limitations employing the following
concepts:

**1. Scoring negative outcomes**


If a person is known to be involved in fraud, we say that there is a negative outcome associated with this
individual. The scores are based on an analysis of fraud cases and their impact from historical data.
From a score of 1 to 10, some negative outcomes can be rated as follows:
image.png

**2. Degree of suspicion (DOS)**

The degree of suspicion (DOS) quantifies our level of suspicion that a person may be involved in fraud. A DOS
value of 0 means that this is a low-risk person and a DOS value of 9 means that this is a high-risk person.
Analysis of historical data shows that professional fraudsters have important positions in their social networks.
To incorporate this:
- Calculate all of the four centrality metrics of each vertex in our network: degree of centrality,
betweenness, closeness and eigenvector.

```python
# Définit la liste des arêtes
edges = [(7, 2), (2, 3), (7, 4), (4, 5), (7, 3), (7, 5), (1, 6), (1, 7), (2, 8), (2, 9)]

# Crée le graph
G = nx.Graph()
G.add_edges_from(edges)

# Définit les nœuds de fraude et de non-fraude
fraud_nodes = [2, 5, 6, 7]
non_fraud_nodes = [1, 4, 3, 8, 9]

# Crée une correspondance des nœuds avec les étiquettes et les scores assignés
node_labels = {}
negative_outcome_scores = {
    'Impersonation': 10,
    'Involvement in credit card theft': 8,
    'Fake Check submission': 7,
    'Criminal record': 6,
    'No record': 0
}
node_labels[2] = ('F', 'Impersonation') # Affectation du score d'usurpation d'identité au nœud de fraude 2
node_labels[6] = ('F', 'Involvement in credit card theft')  # Affectation du score d'implication dans le vol de cartes de crédit au nœud de fraude 6
node_labels[7] = ('F', 'Fake Check submission')  # Affectation du score de soumission de faux chèques au nœud de fraude 7
node_labels[5] = ('F', 'Criminal record')  # Affectation du score de casier judiciaire au nœud de fraude 5

for non_fraud_node in non_fraud_nodes :
  node_labels[non_fraud_node] = ('NF', 'No record')  # Affectation du score d'absence de dossier aux nœuds non-fraudes

# Calcul des métriques de centralité pour chaque nœud
degree_centrality = nx.degree_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)
closeness_centrality = nx.closeness_centrality(G)
eigenvector_centrality = nx.eigenvector_centrality(G)
```

- Take the average of these vertices.
- If a person associated with a vertex is involved in fraud, illustrate this negative outcome by scoring the
person using the pre-determined values shown in the preceding table.
- Multiply the average of the centrality metrics and the negative outcome score to get the value of the
DOS.

```python
# Calcul du DOS pour chaque nœud
normalized_dos_values = {}

for node in G.nodes:
     # Calcul de la moyenne des métriques de centralité
    average_centrality = (degree_centrality[node] + betweenness_centrality[node] + closeness_centrality[node] + eigenvector_centrality[node]) / 4.0

    # Vérifie si le nœud est associé à un cas de fraude connu
    if node in node_labels:
        label, assigned_score = node_labels[node]
        negative_outcome_score = negative_outcome_scores.get(assigned_score, 0)
        print("node :",node," Score :",negative_outcome_score)
    else:
        label = 'NF'
        negative_outcome_score = 0


    if average_centrality != 0:
        # Calcul du DOS
        print("centralité_moyenne :",average_centrality," score_résultat_négatif :",negative_outcome_score)
        dos = average_centrality * negative_outcome_score

    normalized_dos_values[node] = dos

# Trouve la valeur DOS maximale dans le réseau
max_dos_value = max(normalized_dos_values.values())
```

- Finally, normalize the DOS by dividing it by the maximum value of the DOS in the network.


1- Calculate for each vertex, in a given network, its normalized DOS according to the process described
above.

```python
# Normalisation des valeurs DOS en les divisant par la valeur DOS maximale (éviter la division par zéro)
normalized_dos_values = {node: dos / max_dos_value if max_dos_value != 0 else 0 for node, dos in normalized_dos_values.items()}

# Affiche les valeurs DOS normalisées pour chaque nœud
for node, dos in normalized_dos_values.items():
    print(f"Node {node}: Normalized DOS = {dos:.2f}")
```

In order to calculate the DOS of the new node that has been added, the following formula is used:

1- Calculate the DOS of the new node q.

```python
# Ajoute le nœud "q" et les arêtes correspondantes
G.add_node("q")
G.add_edges_from([("q", 1), ("q", 7), ("q", 5)])

# Obtient la liste des voisins de "q"
q_neighbors = list(G.neighbors("q"))

# Initialise une liste pour stocker les DOS des voisins de "q"
neighbor_dos_values = []

# Calcule le DOS pour chaque voisin de "q" en utilisant les valeurs déjà calculées
for neighbor in q_neighbors:
    # Vérifie si le voisin existe dans les valeurs de DOS déjà calculées
    if neighbor in normalized_dos_values:
        dos_neighbor = normalized_dos_values[neighbor]
        # Ajoute le DOS du voisin à la liste
        neighbor_dos_values.append(dos_neighbor)

# Calcule le DOS moyen des voisins de "q"
if neighbor_dos_values:
    average_neighbor_dos = sum(neighbor_dos_values) / len(neighbor_dos_values)
else:
    average_neighbor_dos = 0  # Aucun voisin, donc le DOS moyen est de 0

print(f"DOS moyen des voisins de 'q': {average_neighbor_dos:.2f}")



# Définit les seuils de risque
low_risk_threshold = 0.10
medium_risk_threshold = 0.30
high_risk_threshold = 0.3

# Détermine le niveau de risque en fonction du DOS calculé.
if average_neighbor_dos == 0:
    risk_level = "No Risk"
elif 0 < average_neighbor_dos <= low_risk_threshold:
    risk_level = "Low Risk"
elif low_risk_threshold < average_neighbor_dos <= medium_risk_threshold:
    risk_level = "Medium Risk"
else:
    risk_level = "High Risk"

print(f"Niveau de risque pour le nœud 'q' : {risk_level}")
```
