# **Part 2: Building a simple recommender**

1. Calculate the weighted rating score.

```python
columns = ['title', 'vote_count', 'vote_average', 'runtime']
df = pd.read_csv('movies_metadata.csv', usecols=columns)

# convertion du vote_count en float
df['vote_count'] = df['vote_count'].apply(to_float)

# calcul du mean rating de tout les films le C
C = df['vote_average'].mean()

# calcul du nombre min of votes en utilisant quantile le m
m = df['vote_count'].quantile(0.8)
```

```python
def weighted_rating(row, m=m, C=C):
    v = row['vote_count']
    R = row['vote_average']
    if v >= m and row['runtime'] >= 45 and row['runtime'] <= 300:
        return (v / (v + m)) * R + (m / (v + m)) * C
    else:
        return 0
```

2. Add the score as a feature to the dataframe, by using Apply function.

```python
df['score'] = df.apply(weighted_rating, axis=1)
```

3. Sort the dataframe on the basis of the score just computed.

```python
df = df.sort_values(by='score', ascending=False)
```

4. Output the list of top movies.

```python
top_movies = df[['title', 'score']].head(15)
print("Top Movies:")
print(top_movies)
```

# **Part 3: Implement a user-based collaborative filtering system**

1. Using the data in ratings_small.csv, construct the user rating matrix (where the rows represent the
users, and the columns the movies).

```python
ratings_df = pd.read_csv('ratings_small.csv')

user_rating_matrix = ratings_df.pivot(index='userId', columns='movieId', values='rating')

print("User Rating Matrix:")
print(user_rating_matrix)
```

2. This matrix will be sparse (lot of unknown values), explain why. Propose a way to fill in the missing
values and justify it. Implement it to have a matrix without missing values.

Il y a plein de valeurs inconnues car les utilisateurs pas non pas forcement vue ou noté tout les films existants. On implemente les valeurs manquantes par colonne en donnant la moyenne obtenu par celui-ci car cela permet de mieux comprendre comment la plupart des gens ont aimé chaque film,et non par ligne car si on prenait la moyenne par utilisateur, chaque personne a des goûts différents, ce qui rendrait difficile la comparaison entre les films.

```python
# Calcul de la moyenne de chaque colonnes
column_means = user_rating_matrix.mean()
column_means
```

```python
# On remplis les Valeurs NaN par ce qu'on a obtenu au dessus
user_rating_matrix_avec_moyenne = user_rating_matrix.fillna(column_means)
user_rating_matrix_avec_moyenne
```

3. Calculate the cosine similarity for each pair of users, in order to obtain a user-user similarity matrix.

```python
#Calcul du cosine similarity grace a la matrice contenant les moyennes au lieux du NaN
from sklearn.metrics.pairwise import cosine_similarity

matrice_similarite = cosine_similarity(user_rating_matrix_avec_moyenne)
matrice_similarite = pd.DataFrame(matrice_similarite, index=user_rating_matrix_avec_moyenne.index, columns=user_rating_matrix_avec_moyenne.index)

matrice_similarite
```

4. Implement a function to return the n nearest neighbors of a given user in the sense of the calculated
similarity.

```python
#On prend les valeurs 5 premier id dont les valeurs sont les plus grandes de la ligne du user_id choisi
def get_nearest_neighbors(user_id, n_neighbors=5):

    similarity_scores = matrice_similarite[user_id]
    sorted_users = sorted(enumerate(similarity_scores), key=lambda x: x[1], reverse=True)
    nearestneighbors = [user for user, _ in sorted_users[1:n_neighbors+1]]

    return nearestneighbors

nearest_neighbors = get_nearest_neighbors(user_id=1, n_neighbors=5)
print(f"Le top 5 des voisins les plus proches de l'utilisateur 1 sont {nearest_neighbors}")
```

```python
import pandas as pd
def recommend_movies(user_id, k,n_neighbors):

  ratings_df = pd.read_csv('ratings_small.csv')

  #on recupere la matrice de notation avec les NaN
  rating_matrix = ratings_df.pivot(index='userId', columns='movieId', values='rating')

  #ici on recupere tout les films non vue par user_id
  user_ratings = rating_matrix.loc[user_id].dropna()
  unrated_movie_columns = rating_matrix.columns[~rating_matrix.columns.isin(user_ratings.index)]
  #print(unrated_movie_columns)

  #On recupere les 5 nearest_neighbors de User_id avec la fonction precedente get_nearest_neighbors
  nearest_neighbors = get_nearest_neighbors(user_id, n_neighbors)
  user_rating_matrix_neighbors = rating_matrix.loc[nearest_neighbors]
  #print(user_rating_matrix_neighbors)

  #On fait un filtre on enleve tout les films vue par user_id puis tout ceux qui non pas etait vue par au moins un des plus proche voisin avec ce qu'on a obtenu au dessus
  user_rating_matrix_neighbors = user_rating_matrix_neighbors[unrated_movie_columns]
  user_rating_matrix_neighbors_cleaned = user_rating_matrix_neighbors.dropna(axis=1, how='all')

  #On prend les scores de similarite des voisins
  matrice_similarite = cosine_similarity(user_rating_matrix_avec_moyenne)
  matrice_similarite = pd.DataFrame(matrice_similarite, index=user_rating_matrix_avec_moyenne.index, columns=user_rating_matrix_avec_moyenne.index)
  similarity_scores = matrice_similarite.loc[user_id, nearest_neighbors]
  print(similarity_scores)

  #on multiplie chaque score donner par les plus proches voisins par son score de similarite
  weighted_user_ratings = user_rating_matrix_neighbors_cleaned.mul(similarity_scores, axis=0)
  #print(weighted_user_ratings)

  # Somme des scores de similarité
  column_sums = weighted_user_ratings.sum(axis=0)
  #print(column_sums)


  similarity_scores=similarity_scores.tolist()
  # Créez une liste pour stocker les sommes pondérées des scores de similarité par colonne
  sum_similarity_scores = []


  # On va recuperer la sum de similarité utiliser par colonne
  for column_name, column_data in user_rating_matrix_neighbors_cleaned.iteritems():
    total_similarity_score = sum(
        similarity_scores[i]
        for i, rating in enumerate(column_data) if pd.notna(rating)
    )
    sum_similarity_scores.append(total_similarity_score)

  #print(sum_similarity_scores)
  #print(column_sums)


  # Divise chaque chaque element de column_sums par sum_similarity_scores
  normalized_column_sums = column_sums / sum_similarity_scores
  #print(normalized_column_sums)

  # et on sort par ordre croissant
  sorted_normalized_column_sums = normalized_column_sums.sort_values(ascending=False)

  #print(sorted_normalized_column_sums)

  #Puis on recuperer les 10 premiers film
  top_10_movies = sorted_normalized_column_sums.head(k)

  return top_10_movies

user=1
recommended_movies = recommend_movies(user_id=user, k=50, n_neighbors=5)
print("recommendation des 10 premiers films pour l'User ",user)
print(recommended_movies)
```
